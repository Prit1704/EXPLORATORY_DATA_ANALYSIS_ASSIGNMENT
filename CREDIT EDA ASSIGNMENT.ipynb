```python
plt.figure(figsize = (5,4))
sns.barplot(data = inp1, x = 'CODE_GENDER', y = 'AMT_INCOME_TOTAL')
plt.xlabel("GENDER")
plt.ylabel("Amount Income Total")
plt.show()
```

```python
plt.figure(figsize = (5,4))
plt.scatter(inp1.AMT_INCOME_TOTAL, inp1.AMT_CREDIT)
plt.xlabel("Amount Income")
plt.ylabel("Amount Credit")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (5,4))
sns.barplot(data = inp1, x = 'CODE_GENDER', y = 'AMT_ANNUITY', 
            hue = 'TARGET')
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.barplot(data = inp1, x = 'CODE_GENDER', y = 'AMT_ANNUITY', 
            hue = 'FLAG_OWN_REALTY')
plt.show()
```

```python
inp1.NAME_CONTRACT_TYPE.value_counts().plot.barh()
```

```python
sns.boxplot(data = inp1, x = 'AMT_ANNUITY')
plt.show()
```

```python
res = pd.pivot_table(data = inp1, columns= "NAME_CONTRACT_TYPE", 
               values = 'AMT_INCOME_TOTAL',
              index = 'CODE_GENDER')
res
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(res, cmap = "Reds", annot = True)
plt.show()
```

```python
res = pd.pivot_table(data = inp1, columns= "NAME_CONTRACT_TYPE", 
               values = 'AMT_CREDIT',
              index = 'CODE_GENDER')
res
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(res, cmap = "Greens", annot = True)
plt.show()
```
